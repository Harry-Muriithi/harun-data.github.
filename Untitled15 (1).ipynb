```python
import requests
from tqdm import tqdm
import os

# URLs of the dataset files
urls = [
    "https://data.mendeley.com/public-files/datasets/s9bfhswzjb/files/cbef7f6d-4f45-45f2-b751-192a60f79758/file_downloaded",
    "https://data.mendeley.com/public-files/datasets/s9bfhswzjb/files/13267974-54bf-42b0-87cb-f691b48ba384/file_downloaded"
]

# Directory to save the files
save_dir = "eye_disease_dataset"
os.makedirs(save_dir, exist_ok=True)

# Function to download large files in chunks
def download_file(url, save_path):
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    block_size = 1024 * 1024  # 1MB per chunk

    with open(save_path, "wb") as file, tqdm(
        desc=save_path,
        total=total_size,
        unit="B",
        unit_scale=True,
        unit_divisor=1024
    ) as bar:
        for chunk in response.iter_content(block_size):
            file.write(chunk)
            bar.update(len(chunk))

# Download each file
for i, url in enumerate(urls):
    filename = f"dataset_part_{i+1}.zip"  # Change if files have known extensions
    save_path = os.path.join(save_dir, filename)
    download_file(url, save_path)

print("Download complete!")
```

```python
import zipfile
import tarfile

# Function to extract files
def extract_dataset(file_path, extract_to):
    if file_path.endswith(".zip"):
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)
    elif file_path.endswith(".tar") or file_path.endswith(".tar.gz"):
        with tarfile.open(file_path, 'r:*') as tar_ref:
            tar_ref.extractall(extract_to)

# Extract each downloaded file
for file in os.listdir(save_dir):
    extract_dataset(os.path.join(save_dir, file), save_dir)

print("Extraction complete!")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Define the dataset path
dataset_path = save_dir

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load dataset
dataset = datasets.ImageFolder(root=dataset_path, transform=transform)

# Create a DataLoader
dataloader = DataLoader(dataset, batch_size=32, shuffle=True, num_workers=4)

# Print dataset details
print(f"Loaded {len(dataset)} images from {len(dataset.classes)} classes.")
print(f"Class Names: {dataset.classes}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to display images
def imshow(img):
    img = img.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = std * img + mean
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Get a batch of images
dataiter = iter(dataloader)
images, labels = next(dataiter)

# Show the first image in the batch
imshow(images[0])
print(f"Label: {dataset.classes[labels[0]]}")
```

```python
#calss distribution of original Data

import os
import matplotlib.pyplot as plt
from collections import Counter

dataset_path = "eye_disease_dataset/Original Dataset"

# Count images per class
classes = os.listdir(dataset_path)
class_counts = {cls: len(os.listdir(os.path.join(dataset_path, cls))) for cls in classes}

# Plot class distribution
plt.figure(figsize=(12, 6))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.xticks(rotation=45)
plt.xlabel("Disease Categories")
plt.ylabel("Number of Images")
plt.title("Class Distribution in the Original Dataset")
plt.show()
```

```python
#Images per Class
import random
from PIL import Image

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.ravel()

for i, cls in enumerate(random.sample(classes, 10)):
    img_path = os.path.join(dataset_path, cls, random.choice(os.listdir(os.path.join(dataset_path, cls))))
    img = Image.open(img_path)
    axes[i].imshow(img)
    axes[i].set_title(cls)
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```

```python
#Class Distribution in the Augmented Dataset
# Count images per class in the Augmented Dataset
augmented_dataset_path = "eye_disease_dataset/Augmented Dataset"
augmented_classes = os.listdir(augmented_dataset_path)
augmented_class_counts = {cls: len(os.listdir(os.path.join(augmented_dataset_path, cls))) for cls in augmented_classes}

# Plot class distribution
plt.figure(figsize=(12, 6))
plt.bar(augmented_class_counts.keys(), augmented_class_counts.values(), color='lightcoral')
plt.xticks(rotation=45)
plt.xlabel("Disease Categories")
plt.ylabel("Number of Images")
plt.title("Class Distribution in the Augmented Dataset")
plt.show()

```

```python
# Comparison of Class Distributions: Original vs. Augmented Dataset
import numpy as np

# Create a grouped bar chart
labels = list(class_counts.keys())
original_values = [class_counts[cls] for cls in labels]
augmented_values = [augmented_class_counts.get(cls, 0) for cls in labels]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width/2, original_values, width, label="Original Dataset", color="skyblue")
rects2 = ax.bar(x + width/2, augmented_values, width, label="Augmented Dataset", color="lightcoral")

ax.set_xlabel("Disease Categories")
ax.set_ylabel("Number of Images")
ax.set_title("Comparison of Class Distributions: Original vs. Augmented Dataset")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

plt.show()

```

```python
#Pixel Intensity Distribution
import numpy as np
import cv2

pixel_values = []

# Sample a few images from different classes
for cls in classes[:5]:
    class_folder = os.path.join(dataset_path, cls)
    img_files = os.listdir(class_folder)[:3]  # Sample 3 images per class
    for img_file in img_files:
        img_path = os.path.join(class_folder, img_file)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            pixel_values.extend(img.flatten())

# Plot histogram
plt.figure(figsize=(10, 5))
plt.hist(pixel_values, bins=50, color="purple", alpha=0.7)
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")
plt.title("Pixel Intensity Distribution")
plt.show()
```
