```python
import numpy as np
import os
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, optimizers, callbacks
import tensorflow_probability as tfp
import matplotlib.pyplot as plt

# Parameters
IMAGE_SIZE = 32
N_COMPONENTS = 5
EPOCHS = 10
BATCH_SIZE = 128

# Load and preprocess data
(x_train, _), (_, _) = datasets.fashion_mnist.load_data()

def preprocess(imgs):
    imgs = np.expand_dims(imgs, -1)  # Add channel dimension
    imgs = tf.image.resize(imgs, (IMAGE_SIZE, IMAGE_SIZE)).numpy()
    return imgs.astype(np.uint8)  # Keep as uint8 in [0, 255]

input_data = preprocess(x_train)
```

```python
# Display sample images
def display(images, n=10, save_to=None):
    plt.figure(figsize=(n, 1))
    for i in range(n):
        plt.subplot(1, n, i+1)
        plt.imshow(images[i].squeeze(), cmap="gray")
        plt.axis("off")
    if save_to:
        plt.savefig(save_to)
    plt.show()

display(input_data)
```

```python
# Create PixelCNN distribution
dist = tfp.distributions.PixelCNN(
    image_shape=(IMAGE_SIZE, IMAGE_SIZE, 1),
    num_resnet=1,
    num_hierarchies=2,
    num_filters=32,
    num_logistic_mix=N_COMPONENTS,
    dropout_p=0.3,
)

# Wrap log_prob in a Keras layer
class LogProbLayer(tf.keras.layers.Layer):
    def __init__(self, dist):
        super().__init__()
        self.dist = dist

    def call(self, inputs):
        inputs = tf.cast(inputs, tf.float32)
        return self.dist.log_prob(inputs)
```

```python
# Wrap log_prob and add loss inside a layer
class LogProbLayer(tf.keras.layers.Layer):
    def __init__(self, dist):
        super().__init__()
        self.dist = dist

    def call(self, inputs):
        inputs = tf.cast(inputs, tf.float32)
        log_prob = self.dist.log_prob(inputs)
        self.add_loss(-tf.reduce_mean(log_prob))  # Add loss INSIDE the layer
        return log_prob  # Still return for monitoring if needed

# Build the model
image_input = layers.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 1), dtype=tf.uint8)
log_prob_output = LogProbLayer(dist)(image_input)

pixelcnn = models.Model(inputs=image_input, outputs=log_prob_output)
pixelcnn.compile(optimizer=optimizers.Adam(0.001))
```

```python
# Callback to generate images
class ImageGenerator(callbacks.Callback):
    def __init__(self, num_img):
        self.num_img = num_img

    def generate(self):
        return dist.sample(self.num_img).numpy()

    def on_epoch_end(self, epoch, logs=None):
        generated_images = self.generate()
        output_dir = "./output"
        os.makedirs(output_dir, exist_ok=True)
        path = os.path.join(output_dir, f"generated_img_{epoch:03d}.png")
        display(generated_images, n=self.num_img, save_to=path)

img_generator_callback = ImageGenerator(num_img=5)
tensorboard_callback = callbacks.TensorBoard(log_dir="./logs")
```

```python
# Train the model and store the history object
history = pixelcnn.fit(
    input_data,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    verbose=True,
    callbacks=[tensorboard_callback, img_generator_callback],
)

# Plot the training loss
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss (Negative Log-Likelihood)')
plt.title('PixelCNN Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Final image generation
generated_images = img_generator_callback.generate()
display(generated_images, n=5)
```
