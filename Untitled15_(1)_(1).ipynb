```python
import requests
from tqdm import tqdm
import os

# URLs of the dataset files
urls = [
    "https://data.mendeley.com/public-files/datasets/s9bfhswzjb/files/cbef7f6d-4f45-45f2-b751-192a60f79758/file_downloaded",
    "https://data.mendeley.com/public-files/datasets/s9bfhswzjb/files/13267974-54bf-42b0-87cb-f691b48ba384/file_downloaded"
]

# Directory to save the files
save_dir = "eye_disease_dataset"
os.makedirs(save_dir, exist_ok=True)

# Function to download large files in chunks
def download_file(url, save_path):
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    block_size = 1024 * 1024  # 1MB per chunk

    with open(save_path, "wb") as file, tqdm(
        desc=save_path,
        total=total_size,
        unit="B",
        unit_scale=True,
        unit_divisor=1024
    ) as bar:
        for chunk in response.iter_content(block_size):
            file.write(chunk)
            bar.update(len(chunk))

# Download each file
for i, url in enumerate(urls):
    filename = f"dataset_part_{i+1}.zip"  # Change if files have known extensions
    save_path = os.path.join(save_dir, filename)
    download_file(url, save_path)

print("Download complete!")
```

```python
import zipfile
import tarfile

# Function to extract files
def extract_dataset(file_path, extract_to):
    if file_path.endswith(".zip"):
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)
    elif file_path.endswith(".tar") or file_path.endswith(".tar.gz"):
        with tarfile.open(file_path, 'r:*') as tar_ref:
            tar_ref.extractall(extract_to)

# Extract each downloaded file
for file in os.listdir(save_dir):
    extract_dataset(os.path.join(save_dir, file), save_dir)

print("Extraction complete!")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Define the dataset path
dataset_path = save_dir

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load dataset
dataset = datasets.ImageFolder(root=dataset_path, transform=transform)

# Create a DataLoader
dataloader = DataLoader(dataset, batch_size=32, shuffle=True, num_workers=4)

# Print dataset details
print(f"Loaded {len(dataset)} images from {len(dataset.classes)} classes.")
print(f"Class Names: {dataset.classes}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to display images
def imshow(img):
    img = img.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = std * img + mean
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Get a batch of images
dataiter = iter(dataloader)
images, labels = next(dataiter)

# Show the first image in the batch
imshow(images[0])
print(f"Label: {dataset.classes[labels[0]]}")
```

```python
#Pixel Intensity Distribution
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define dataset path
dataset_path = "eye_disease_dataset/Original Dataset"  # Update if needed

# Get list of disease categories
classes = sorted(os.listdir(dataset_path))

# Initialize lists for RGB values (Sample fewer images)
r_values, g_values, b_values = [], [], []

# Limit number of images to process (e.g., 1 per class)
num_samples_per_class = 1

for cls in classes[:5]:  # Sample first 5 classes
    class_folder = os.path.join(dataset_path, cls)
    img_files = os.listdir(class_folder)[:num_samples_per_class]  # Take 1 image per class

    for img_file in img_files:
        img_path = os.path.join(class_folder, img_file)
        img = cv2.imread(img_path)

        if img is not None:
            img = cv2.resize(img, (100, 100))  # Reduce resolution to lower RAM usage
            r_values.extend(img[:, :, 2].flatten()[::10])  # Sample every 10th pixel
            g_values.extend(img[:, :, 1].flatten()[::10])
            b_values.extend(img[:, :, 0].flatten()[::10])

# Plot histograms with reduced bins (Lower RAM usage)
plt.figure(figsize=(10, 5))
plt.hist(r_values, bins=30, color="red", alpha=0.6, label="Red Channel")
plt.hist(g_values, bins=30, color="green", alpha=0.6, label="Green Channel")
plt.hist(b_values, bins=30, color="blue", alpha=0.6, label="Blue Channel")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")
plt.title("Optimized Color Intensity Distribution")
plt.legend()
plt.show()
```

```python
#Edge Density Distribution
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define dataset path
dataset_path = "eye_disease_dataset/Original Dataset"

# Get list of classes
classes = sorted(os.listdir(dataset_path))

edge_densities = []

# Limit number of images (1 per class)
num_samples_per_class = 1

for cls in classes[:5]:
    class_folder = os.path.join(dataset_path, cls)
    img_files = os.listdir(class_folder)[:num_samples_per_class]  # Take 1 image per class

    for img_file in img_files:
        img_path = os.path.join(class_folder, img_file)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale

        if img is not None:
            img = cv2.resize(img, (100, 100))  # Reduce size
            edges = cv2.Canny(img, 50, 150)  # Detect edges
            edge_density = np.sum(edges) / (100 * 100)  # Normalize edge count
            edge_densities.append(edge_density)

# Plot histogram of edge density values
plt.figure(figsize=(8, 4))
plt.hist(edge_densities, bins=10, color="blue", alpha=0.7)
plt.xlabel("Edge Density")
plt.ylabel("Frequency")
plt.title("Optimized Edge Density Distribution")
plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Define dataset path
dataset_path = "eye_disease_dataset/Original Dataset"

# Get list of classes
classes = sorted(os.listdir(dataset_path))

# Number of images to display
num_images = 5
sample_images = []

# Select random images from different classes
for cls in random.sample(classes, min(num_images, len(classes))):
    class_folder = os.path.join(dataset_path, cls)
    img_files = os.listdir(class_folder)

    if img_files:
        img_path = os.path.join(class_folder, random.choice(img_files))
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB
            sample_images.append((cls, img))

# Plot images
plt.figure(figsize=(14, 6))  # Increase figure size for better spacing
for i, (cls, img) in enumerate(sample_images):
    plt.subplot(1, len(sample_images), i + 1)
    plt.imshow(img)
    plt.title(cls, fontsize=10)  # Reduce font size slightly
    plt.axis("off")

plt.suptitle("Sample Images Before Preprocessing", fontsize=14, y=1.05)  # Adjust title position
plt.subplots_adjust(wspace=0.3)  # Add space between images
plt.show()
```

```python
#Data Processing
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet101, VGG19
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
# Build ResNet-101 Model

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Define dataset path
dataset_path = "eye_disease_dataset/Original Dataset"

# Image size required for ResNet-101 & VGG-19
IMG_SIZE = (224, 224)
BATCH_SIZE = 32  # Optimal batch size for training

# Data augmentation for training set
train_datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalize pixel values
    rotation_range=15,  # Rotate images randomly
    width_shift_range=0.1,  # Horizontal shift
    height_shift_range=0.1,  # Vertical shift
    zoom_range=0.2,  # Zoom into images
    horizontal_flip=True,  # Flip images
    validation_split=0.2  # 20% validation data
)

# Training dataset
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training"
)

# Validation dataset
val_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation"
)
```

```python
from tensorflow.keras.applications import ResNet101
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout

# Load ResNet-101 without the top layer
base_model = ResNet101(weights="imagenet", include_top=False, input_shape=(224, 224, 3))

# Freeze initial layers (keep pre-trained weights)
for layer in base_model.layers[:-20]:  # Fine-tune last 20 layers
    layer.trainable = False

# Create custom classification model
model_resnet = Sequential([
    base_model,
    Flatten(),
    Dropout(0.5),  # Regularization
    Dense(512, activation="relu"),
    Dense(len(train_generator.class_indices), activation="softmax")  # Output layer
])

# Compile model
model_resnet.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                     loss="categorical_crossentropy",
                     metrics=["accuracy"])

model_resnet.summary()
```

```python
#Train ResNet-101 Model
# Train model
history_resnet = model_resnet.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,  # Adjust if accuracy is low
    verbose=1
)
```

```python
from tensorflow.keras.applications import VGG19

# Load VGG-19 model
base_model_vgg = VGG19(weights="imagenet", include_top=False, input_shape=(224, 224, 3))

# Fine-tune last 10 layers
for layer in base_model_vgg.layers[:-10]:
    layer.trainable = False

# Define custom model
model_vgg = Sequential([
    base_model_vgg,
    Flatten(),
    Dropout(0.5),
    Dense(512, activation="relu"),
    Dense(len(train_generator.class_indices), activation="softmax")
])

# Compile model
model_vgg.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss="categorical_crossentropy",
                  metrics=["accuracy"])

# Train model
history_vgg = model_vgg.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    verbose=1
)
```

```python
# Evaluate ResNet-101
resnet_acc = model_resnet.evaluate(val_generator)[1] * 100
print(f"ResNet-101 Accuracy: {resnet_acc:.2f}%")

# Evaluate VGG-19
vgg_acc = model_vgg.evaluate(val_generator)[1] * 100
print(f"VGG-19 Accuracy: {vgg_acc:.2f}%")
```

```python
#Plot Training & Validation Accuracy
import matplotlib.pyplot as plt

# Plot training and validation accuracy
plt.figure(figsize=(10, 5))
plt.plot(history_resnet.history['accuracy'], label='ResNet-101 Training Accuracy', color='blue')
plt.plot(history_resnet.history['val_accuracy'], label='ResNet-101 Validation Accuracy', color='blue', linestyle='dashed')
plt.plot(history_vgg.history['accuracy'], label='VGG-19 Training Accuracy', color='red')
plt.plot(history_vgg.history['val_accuracy'], label='VGG-19 Validation Accuracy', color='red', linestyle='dashed')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid()
plt.show()
```

```python
#Plot Training & Validation Loss
# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(history_resnet.history['loss'], label='ResNet-101 Training Loss', color='blue')
plt.plot(history_resnet.history['val_loss'], label='ResNet-101 Validation Loss', color='blue', linestyle='dashed')
plt.plot(history_vgg.history['loss'], label='VGG-19 Training Loss', color='red')
plt.plot(history_vgg.history['val_loss'], label='VGG-19 Validation Loss', color='red', linestyle='dashed')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid()
plt.show()
```
