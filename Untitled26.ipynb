```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Load and preprocess CIFAR-10 data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = y_train.flatten(), y_test.flatten()

# Define a helper to plot training curves
```

```python
# Define a helper to plot training curves
def plot_history(history, title):
    plt.figure(figsize=(12, 4))

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()
```

```python
# Build a CNN model with configurable stride
def build_model(stride=2):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), strides=stride, activation='relu', input_shape=(32, 32, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# Model A: Baseline (stride=2)
model_stride2 = build_model(stride=2)
history1 = model_stride2.fit(x_train, y_train, epochs=10, validation_split=0.2)
model_stride2.save("cnn_stride2.h5")

```

```python
# Model B: Custom stride ( stride=3)
model_stride3 = build_model(stride=3)
history2 = model_stride1.fit(x_train, y_train, epochs=10, validation_split=0.2)
model_stride3.save("cnn_stride3.h5")
```

```python
# Plot training progress
plot_history(history1, "Stride=2")
plot_history(history2, "Stride=3")
```

```python
# Evaluate both models
print("Evaluating model with stride=2:")
loss2, acc2 = model_stride2.evaluate(x_test, y_test, verbose=2)
print(f"Test Accuracy: {acc2:.4f}")

print("\nEvaluating model with stride=3:")
loss1, acc1 = model_stride1.evaluate(x_test, y_test, verbose=2)
print(f"Test Accuracy: {acc1:.4f}")
```

```python
# Confusion matrix function
def show_conf_matrix(model, title):
    y_pred = np.argmax(model.predict(x_test), axis=1)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(10))
    disp.plot(cmap=plt.cm.Blues)
    plt.title(title)
    plt.show()

# Show confusion matrices
show_conf_matrix(model_stride2, "Confusion Matrix - Stride=2")
show_conf_matrix(model_stride1, "Confusion Matrix - Stride=3")
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Class names for CIFAR-10
class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']

def show_predictions(model, x_data, y_data, stride_label, num_images=10):
    plt.figure(figsize=(15, 5))
    indices = random.sample(range(len(x_data)), num_images)

    for i, idx in enumerate(indices):
        img = x_data[idx]
        true_label = class_names[int(y_data[idx])]
        pred_probs = model.predict(img.reshape(1, 32, 32, 3), verbose=0)
        pred_label = class_names[np.argmax(pred_probs)]

        plt.subplot(2, num_images // 2, i + 1)
        plt.imshow(img)
        plt.axis('off')

        color = "green" if pred_label == true_label else "red"
        plt.title(f"{stride_label}\nPred: {pred_label}\nTrue: {true_label}", color=color)

    plt.suptitle(f"Model Predictions ({stride_label})", fontsize=16)
    plt.tight_layout()
    plt.show()

# Show predictions for both models
print("🎯 Predictions from Model with stride=2:")
show_predictions(model_stride2, x_test, y_test, "Stride 2")

print("🎯 Predictions from Model with stride=3:")
show_predictions(model_stride3, x_test, y_test, "Stride 3")
```
